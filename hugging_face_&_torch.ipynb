```python
from transformers import pipeline

# Load the text generation model
text_generator = pipeline("text-generation")
text_generator
```

```python
print("Creative Writing Assistant: Welcome! I'm here to help you craft amazing stories. Type 'exit' to end the session.")

while True:
    user_prompt = input("You: ")
    if user_prompt.lower() == "exit":
        print("Creative Writing Assistant: Goodbye! Have a great day!")
        break
    story = text_generator(user_prompt, max_length=100, num_return_sequences=1, do_sample=True, temperature=0.7)
    generated_story = story[0]['generated_text'].strip()
    print("Creative Writing Assistant:")
    print(generated_story)
    print("\n" + "-" * 50 + "\n")
```
